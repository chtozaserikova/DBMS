Формат вьюхи с расчетными данными: date - user_id - session_id - session_duration,
где date и user_id - поля из сырых данных
session_id требуется сгенерить по бизнес правилу (Сессией называется активность пользователя, в которой между последовательными действиями проходит менее 15 мин)
session_duration время между первым и последним ивентом в сессии

Результаты работы: Создать архив  с файлами: контактами текстовый файл с кодом запросов: show create table_name (таблица для сырых данных из csv) show create view_name (создание вьюхи с расчетными данными по бизнес условиям)

```python
import pandas as pd
import numpy as np
```

```python
testovoe = pd.read_csv('test_data.csv', names = ['user_id', 'event_time', 'event_id'])
testovoe = testovoe.sort_values('user_id')
testovoe
```

```python
testovoe['event_time']= pd.to_datetime(testovoe['event_time'])
testovoe["id"] = testovoe.index + 1
# testovoe = testovoe.sort_values('id')
testovoe['diff'] = testovoe.groupby('user_id')['event_time'].diff(1)

testovoe[1000:1020]
```

```python
sessions_testovoe = testovoe[(testovoe['diff'].isnull()) | (testovoe['diff'] > '900 seconds')]
sessions_testovoe['session_id'] = sessions_testovoe['id']
sessions_testovoe
```

```python
sessions_testovoe.loc[sessions_testovoe['id'] == 2]
```

```python
testovoe = testovoe.sort_values('id')
sessions_testovoe = sessions_testovoe.sort_values('id')
testovoe = pd.merge_asof(testovoe,sessions_testovoe[['id','user_id','session_id']],on='id',by='user_id')
```

```python
testovoe['session_duration'] = testovoe.groupby(['session_id'])['event_time'].diff(1)
```

```python
print(testovoe['user_id'].nunique()) # Количество пользователей
print(testovoe['session_id'].nunique()) # Количество сессий
print(testovoe['id'].count()) # Количество просмотров страниц (событий)
```
